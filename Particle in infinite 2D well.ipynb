# Solving the Schrodinger equation for a particle in a 2D infinite well

In this Notebook we look at plotting solutions of the Schroedinger equation for a particle in a 2D infintie square well. The time-dependent Schroedinger equation (TDSE) can be written as:

$i\hbar\frac{\partial\Psi}{\partial t} = \big(-\frac{\hbar^2}{2m}\nabla+U\big)\Psi$

In 2D, $\Psi$ can be written as a function of the spatial coordinates $x$ and $y$, and the time $t$: 
$\Psi(x,y,t)=X(x)Y(y)F(t)$

Our TDSE now becomes

$i\hbar XY \frac{\partial F}{\partial t}=-\frac{\hbar^2}{2m}\bigg(YF\frac{\partial ^2 X}{\partial x^2}+XF\frac{\partial^2 Y}{\partial y^2}\bigg)+XYFU$

Now we divide both sides by $XYF$:

$i\hbar \frac{1}{F}\frac{\partial F}{\partial t} = -\frac{\hbar^2}{2m}\big(\frac{1}{X}\frac{\partial^2 X}{\partial x^2}+\frac{1}{Y}\frac{\partial^2 Y}{\partial y^2}\big)+U$

On the LHS we have a partial derivative w.r.t. time, and on the RHS we have partial derivatives w.r.t. the spatial coordinates. Hence, both sides must be equal to a constant, which we call $E$.

Starting with the LHS:

$i\hbar \frac{1}{F}\frac{\partial F}{\partial t}=E \rightarrow \frac{\partial F}{\partial t}= \frac{-iF}{\hbar}E \rightarrow F=e^\frac{-iEt}{\hbar}$ 


On the RHS: 

$-\frac{\hbar^2}{2m}\big(\frac{1}{X}\frac{\partial^2 X}{\partial x^2}+\frac{1}{Y}\frac{\partial^2 Y}{\partial y^2}\big) = E$

(We take $U=0$ inside the well). Thus

$-\frac{\hbar^2}{2m}\frac{1}{X}\frac{\partial^2 X}{\partial x^2} = E_x$ and $-\frac{\hbar^2}{2m}\frac{1}{Y}\frac{\partial^2 Y}{\partial y^2}=E_y$, where $E=E_x+E_y$

Let $k_x=\sqrt\frac{2mE_x}{\hbar^2}$ and $k_y=\sqrt\frac{2mE_y}{\hbar^2}$

We thus have to solve equations of the form $\frac{\partial^2 X}{\partial x^2}=k_x^2X$

We guess a solution for $X$: 
$X(x)=A\sin(k_x x)+B\cos(k_x x)$

Using the boundary conditions that $X(0)=X(a)=0$, we get that $B=0$ and $k_x a = n \pi$. $A$ is, by normalization, $\sqrt \frac{2}{a}$. Here $a$ (and $b$) are the sizes of the sides of the well. 

The same method can be used to solve for $Y$. The total wavefunction $\Psi$ is then:

$\Psi = \sqrt \frac{2}{a}\sin\frac{n_x\pi x}{a}\sqrt \frac{2}{b}\sin\frac{n_y \pi y}{b}e^\frac{-iEt}{\hbar}$

Below we plot this function (and its square) in a contour plot, for any value of $n_x$ and $n_y$.

```python
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
a = 1 #nm
b = 1
hc = 1240
m = 511000

pi = np.pi
nx = 2
ny = 2
Ex = nx*hc**2/(8*m*a**2)
Ey = ny*hc**2/(8*m*b**2)
E = Ex + Ey
print("nx,ny= ",nx,ny,"E = ",E," eV")

x = np.arange(0,a,0.01)
y = np.arange(0,b,0.01)

xp,yp = np.meshgrid(x,y)
psi = (2/np.sqrt(a*b))*np.sin(nx*pi*xp/a)*np.sin(ny*pi*yp/b) 
psi_sq = psi**2
plt.contourf(xp,yp,psi,levels=50)
plt.colorbar(label="psi")
plt.xlabel("x [nm]")
plt.ylabel("y [nm]")
plt.show()

plt.contourf(xp,yp,psi_sq,levels=50)
plt.colorbar(label="psi")
plt.xlabel("x [nm]")
plt.ylabel("y [nm]")
plt.show()
```
