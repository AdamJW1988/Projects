**Two Bulbs and One Capacitor**

We want to find how the currents in a circuit (see image below) change with time. 

Cap_2bulbs.png

Here we follow the loop rule (sum of V around any loop must be zero, and current must be conserved). Starting with loop rule

$I_1 = I_2+I_3$

For the voltages

$V_b-I_3R-I_1R=0$ (loop 1) \\
$-Q/C+I_3R = 0$ (loop 2)

Now we have 3 equations for 3 unknowns.
At $t=0$, we have no charge on capacitor, so 
$I_3=0$, so there is no current going through first bulb. If $I_3 = 0$, then with loop 1 $I_1 = V_b/R$. \\
As time goes on, the capacitor will build up a charge, and thus there will be a non zero $I_3$. This means that the current $I_1$ going through the second bulb must decrease (see loop 2).

$I_2 = \frac{Q}{t} \rightarrow I_1 -I_3 = \frac{Q}{t}$ 

$I_3 = Q/RC \rightarrow I_1-Q/RC = Q/t$

$Q=(I_1-Q/RC)t$

Remembering that $I_1 = \frac{V_b-I_3R}{R}=\frac{V_b-\frac{Q}{RC}R}{R}=\frac{V_b-Q/C}{R}=\frac{V_bC-Q}{RC}$

The initial conditions are  

$Q=0$ 

$I_1 = V_b/2R$

We can solve this using a numerical calculation with Python. The parameters we will use are

$V_b = 3 V$ 

$R = 10 \Omega $

$C = 1 F$

$t = 0.001$





```python
import numpy as np
import matplotlib.pyplot as plt

Vb = 3
R = 10
C = 1
t = 0
dt = 0.001 

Q = 0
I1 = Vb/(2*R)

xcol = []
ycol_I1 = []
ycol_I2 = []
ycol_I3 = []
ycol_Q = []

while t<25:
  dQ = (I1-Q/(R*C))*dt
  Q += dQ
  I3 = Q/(R*C)
  I1 = (Vb-I3*R)/R
  I2 = I1-I3
  t += dt
  ycol_I1.append(I1)
  ycol_I2.append(I2)
  ycol_I3.append(I3)
  ycol_Q.append(Q)
  xcol.append(t)

plt.plot(xcol,ycol_I1,label = 'I1')
plt.plot(xcol,ycol_I2,label = 'I2')
plt.plot(xcol,ycol_I3, label = 'I3')
plt.xlabel('t[s]')
plt.ylabel('I[A]')
plt.title('Numerical solution of currents I1, I2 and I3')
plt.legend()
plt.show()

plt.plot(xcol,ycol_Q)
plt.xlabel('t[s]')
plt.ylabel('Q[C]')
plt.title('Numerical solution of charge Q(t) on capacitor')
plt.show()
```
