# Wave Interference

Suppose we drop a pebble in a pond, and waves ripple outwards. We can model this simply with a sine wave spreading outwards in a circle. If the center of the circle is at $x_1, y_1$ then the distance $r_1$ to the center from a point $x,y$ is

$r_1 = \sqrt{(x-x_1)^2+(y-y_1)^2} $

and the height of the wave at that point is defined as

$\beta _1(x.y) = \beta_0 \sin(kr_1)$

where $\beta_0$ is the amplitude of the waves and $k$ is the wavevector.If we drop a second pebble in the pond at a different location $x_2,y_2$, then

$\beta_2(x,y) = \beta_0\sin(kr_2)$ with $r_2 = \sqrt{(x-x_2)^2+(y-y_2)^2}$

Assuming the waves add linearly, which is reasonable if the waves are not too big, we get

$\beta(x,y) = \beta_0\sin(kr_1)+\beta_1\sin(kr_2)$

Suppose the wavelength of the wave is $\lambda = 5$ cm, the amplitude 1 cm, and the centers of the circles 20 cm apart. The following code will calculate the height of the waves over a 1 m square region of the water. To make the image, we create an array of values representing the height $\beta$ at a grid of points and then use that array to make a density plot. Here we use a grid og 500x500 points to cover the 1 m square, making the seperation of grid points 100/500 = 0.2 cm. 



```python
# This code simulates wave interference

from math import sqrt,sin,pi
from numpy import empty
from pylab import imshow,gray,show

wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 20.0 # seperation of centers in cm
side = 100.0 # side of square in cm
points = 500 # number of grid points along each side
spacing = side/points #spacing of points in cm

# calculate the position of the centers of the circles
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# make empty array to store heights
xi = empty([points,points],float)

# calculate values in array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*sin(k*r1)+xi0*sin(k*r2)

# make plot
imshow(xi,origin="lower",extent=[0,side,0,side])
gray()
show()
```

We can also amend the code to see what happens for different wavevectors. For instance, lets say $k_1 = 2k_2$, which means $\lambda_1 = \frac{1}{2}\lambda_2$

```python
from math import sqrt,sin,pi
from numpy import empty
from pylab import imshow,gray,show

wavelength_1 = 5.0
wavelength_2 = 10
k_1 = 2*pi/wavelength_1
k_2 = 2*pi/wavelength_2
xi0 = 1.0
separation = 20.0 # seperation of centers in cm
side = 100.0 # side of square in cm
points = 500 # number of grid points along each side
spacing = side/points #spacing of points in cm

# calculate the position of the centers of the circles
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# make empty array to store heights
xi = empty([points,points],float)

# calculate values in array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*sin(k_1*r1)+xi0*sin(k_2*r2)

# make plot
imshow(xi,origin="lower",extent=[0,side,0,side])
gray()
show()
```
